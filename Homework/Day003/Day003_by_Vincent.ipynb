* 根據範例檔案的結果：
    1. 請問高雄市有多少地區有溫度資料？
    2. 請取出每一個地區所記錄的第一個時間點跟溫度
    3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
import xmltodict

# 存取檔案
with open('./data/64_72hr_CH.xml',"r",encoding="utf-8") as fd:
    doc = dict(xmltodict.parse(fd.read()))
    locationsName = doc['cwbopendata']['dataset']['locations']['location']
    k=0
    for i in locationsName:
        k+=1
        print(i['locationName'])
    print('高雄市有多少地區有溫度資料:',k)
```

```python
for location in locationsName:
    print(location['locationName'])
    for weatherElement in location['weatherElement']:
        if weatherElement['description'] == '溫度':
            print(weatherElement['time'][0]['dataTime'])
            print(weatherElement['time'][0]['elementValue']['value'])
```

```python
first = locationsName[0]
print(first['locationName'])
for weatherElement in first['weatherElement']:
    if weatherElement['description'] == '溫度':
        time_points = weatherElement['time'] 
        for time in time_points:
            print(time['dataTime'])
            print(time['elementValue']['value'])
```
