### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異


#### 。r.text是純文字的資料型態，不方便取用資料
#### 。json.load(r.text)讀出來是json格式，取用資料上較方便

```python
import requests
import json


r = requests.get('https://api.github.com/events')
json=json.loads(r.text)
print(r.text)
print('json:')
print(json)
```
