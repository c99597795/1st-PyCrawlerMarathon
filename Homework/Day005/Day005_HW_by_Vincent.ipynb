## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
#這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
Fp= requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = r.text
responsef = Fp.text
data=json.loads(response)
dataf=json.loads(responsef)
print('回傳資料數:',len(data))
print('回傳欄位:')
for i in data[0].keys() :
    print(i)
    

```

```python
#取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print('標題:\n',d['title'])
    print('貼文時間:',d['createdAt'])
    print('留言人數:',d['commentCount'])
    print('按讚人數:',d['likeCount'],'\n')
    
```

```python
#計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
p=0 #熱門文章數量
fp=0 #非熱門文章數量
pComment=0
fpComment=0
pLike=0
fpLike=0
for d in data:
    p+=1#熱門文章數
for d in data:
    pLike+=d['likeCount']#熱門文章平均按讚人數
for d in data:
    pComment+=d['commentCount'] #熱門文章平均留言人數
for d in dataf:
    fp+=1
for d in dataf:
    fpLike+=d['likeCount']#非熱門文章平均按讚人數
for d in dataf:
    fpComment+=d['commentCount']#非熱門文章平均留言人數
pLike=(pLike//p)
pComment=(pComment//p)
fpLike=(fpLike//p)
fpComment=(fpComment//p)

print('熱門文章數量',p)
print('非熱門文章數量',fp)
print('熱門文章平均按讚人數',pLike)
print('熱門文章平均留言人數',pComment)
print('非熱門文章平均按讚人數',fpLike)
print('非熱門文章平均留言人數',fpComment)

```
